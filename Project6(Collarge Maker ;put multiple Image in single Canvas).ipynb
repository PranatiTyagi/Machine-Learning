```python
image_path=[]
image_name=[]
for i in image:
    image_name.append(i)
    image_path.append(os.path.join(path,i))
print(image_name)
print(image_path)
```

```python
size=100
Canvas=np.zeros((2*size,4*size,3),'uint8')
images=[np.array(Image.open(file))[:size,:size] for file in image_path]
image_path[0]
```

```python
k=0
for row in range(2):
    for column in range(4):
        Canvas[(row*size):(row+1)*size,(column*size):(column+1)*size]=images[k]
        k+=1
plt.imshow(Canvas)
```
