```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
model=cv.CascadeClassifier("haar_frontal.xml")#model:lenit5,gans(video) loading pretrained model
```

```python
cam=cv.VideoCapture(0)
model=cv.CascadeClassifier("haar_frontal.xml")
while True:
    _,frame=cam.read()
    fr=cv.flip(frame,+1)
    gray=cv.cvtColor(fr,cv.COLOR_BGR2GRAY)
    
    try:
        faces=model.detectMultiScale(gray,1.3,5)
        for face in faces:
            x,y,w,h=face
            cv.rectangle(fr,(x,y),(x+w,y+h),(255,255,255),2)    
    except:
        print("error")
    cv.imshow("Fce Detection",fr)
    key=cv.waitKey(1)
    if key==ord('q'):
        cv.destroyAllWindows()
        break
cam.release()    
plt.imshow(fr)   
```
